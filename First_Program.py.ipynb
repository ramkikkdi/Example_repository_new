```python
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
```

```python
from tkinter import ttk
```

```python
## Windows dpi resolution increase
from ctypes import windll
windll.shcore.SetProcessDpiAwareness(1)
```

```python
## This is to create a base window for the application and so we can build the widgets upon this object
root = tk.Tk()

### The below is the width x height of the window
root.geometry("2000x1000")
##root.resizable(False,False)

## This is for the title page
root.title("Title page")

## To create a tabs in the python tkinter window with the help of themed tkiner widgets
notebook = ttk.Notebook(root)
notebook.pack(padx= 20, pady=20, expand = True)

## To create a frame for packing the notebooks
frame1 = ttk.Frame(notebook, width=1200, height=1000)
frame2 = ttk.Frame(notebook, width=1200, height=1000)

## To pack these frames into the window
frame1.pack(fill='both', expand=True)
frame2.pack(fill='both', expand=True)

## To add the frames to the notebook
notebook.add(frame1, text='Temperature')
notebook.add(frame2, text='Torque')

## This is to execute the main loop
root.mainloop()
```
